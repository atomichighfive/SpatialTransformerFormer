```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
test_history = pd.read_csv("test_history.csv")
train_history = pd.read_csv("train_history.csv")

train_history["loss"] = train_history["loss"] / train_history["loss"].max()
test_history["loss"] = test_history["loss"] / test_history["loss"].max()

test_ax = train_history[["loss", "accuracy"]].plot()
plt.title("Train")
plt.grid()
plt.show()
test_history[["loss", "accuracy"]].plot()
plt.title("Test")
plt.grid()
plt.show()
```
