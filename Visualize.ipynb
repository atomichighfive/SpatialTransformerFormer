```python
import pandas as pd
from matplotlib import pyplot as plt
from pathlib import Path
```

```python
def plot_history(checkpoint_dirs):
    test_history = pd.concat([pd.read_csv(cpd / "test_history.csv") for cpd in checkpoint_dirs], axis=0).reset_index()
    train_history = pd.concat([pd.read_csv(cpd / "train_history.csv") for cpd in checkpoint_dirs], axis=0).reset_index()

    train_history["loss"] = train_history["loss"] / train_history["loss"].max()
    test_history["loss"] = test_history["loss"] / test_history["loss"].max()

    test_ax = train_history[["loss", "accuracy"]].plot()
    plt.title("Train")
    plt.grid()
    plt.show()
    test_history[["loss", "accuracy"]].plot()
    plt.title("Test")
    plt.grid()
    plt.show()
```

# Examine full training history

```python
plot_history(
    [
        Path("model_checkpoints/run_2.0/"),
        #Path("model_checkpoints/run_2.1/"),  # Skipped because 2.2 contains this.
        Path("model_checkpoints/run_2.2/"),
        Path("model_checkpoints/run_2.3_reinitialize_loc_layers/"),
    ]
)
```

# Look at only training since loc layers were reinitialized

```python
plot_history(
    [
        Path("model_checkpoints/run_2.3_reinitialize_loc_layers/"),
    ]
)
```
